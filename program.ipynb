```python
import pandas as pd
import json
```

```python
kaggle = pd.read_csv("Production5.csv")
kaggle.head(5)
```

```python
production_1 = kaggle[['title', 'revenue']]
production_1.head(3)
```

```python
with open('production_n_1.csv', 'a') as f:
    c1 = production_1.to_csv(f, index=True, header=True)
```

```python
def extractJsonObj(obj):
    temp = []
    for subline in obj:
        temp.append(subline['name'])
    return temp
    
kaggle['production_company_names'] = kaggle['production_companies'].apply(lambda x: extractJsonObj(json.loads(x)))
kaggle['production_country_names'] = kaggle['production_countries'].apply(lambda x: extractJsonObj(json.loads(x)))
```

```python
kaggle.head(3)
```

```python
p_company = kaggle[['title', 'production_company_names']]
```

```python
p_company.head(5)
```

```python
atomic_company = p_company.production_company_names.apply(pd.Series) \
    .merge(p_company, left_index = True, right_index = True) \
    .drop(["production_company_names"], axis = 1) \
    .melt(id_vars = ['title'], value_name = "production_company_name") \
    .drop("variable", axis = 1) \
    .dropna()
```

```python
atomic_company.head(5)
```

```python
with open('pcompany_n_1.csv', 'a') as f:
    c1 = atomic_company.to_csv(f, index=True, header=True)
```

```python
p_country = kaggle[['title', 'production_country_names']]
```

```python
p_country.head(5)
```

```python
atomic_country = p_country.production_country_names.apply(pd.Series) \
    .merge(p_country, left_index = True, right_index = True) \
    .drop(["production_country_names"], axis = 1) \
    .melt(id_vars = ['title'], value_name = "production_country_name") \
    .drop("variable", axis = 1) \
    .dropna()
```

```python
atomic_country.head(5)
```

```python
with open('pcountry_n_1.csv', 'a') as f:
    c1 = atomic_country.to_csv(f, index=True, header=True)
```

```python
user_info = pd.read_csv("tweet_user_info1.csv")
user_info
```

```python
tweet_info = pd.read_csv("Twitter_consumer_1.csv")
tweet_info.head(5)
```

```python
tweet_merge = pd.merge(tweet_info, user_info, on =['User_id','User_id'])
tweet_merge.head(5)
```

```python
tweet_user_info = tweet_merge[['Name','User_id','Followers','title','User_name']]
tweet_user_info.head(5)
```

```python
tweet_user_details = tweet_user_info.loc[tweet_user_info['Name'] != tweet_user_info['Name'].shift()]
tweet_user_details
```

```python
with open('tweet_user_details1.csv', 'a') as f:
    s1 = tweet_user_details.to_csv(f, index = True, header=True)
```
